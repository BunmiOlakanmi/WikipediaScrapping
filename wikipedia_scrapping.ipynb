```python
# Custom Class which confirms datatype is readable by MongoBD
# and if test fail, convert to readable variable
import json

class CustomEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, numpy.integer):
            return int(obj)
        elif isinstance(obj, numpy.floating):
            return float(obj)
        elif isinstance(obj, numpy.ndarray):
            return obj.tolist()
        else:
            return super(CustomEncoder, self).default(obj)
```

```python
#Collection        
df_dict=df_3.to_dict("records")
df_3_json = json.dumps(df_dict,cls=CustomEncoder)
df_3_final  = json.loads(df_3_json)
```
